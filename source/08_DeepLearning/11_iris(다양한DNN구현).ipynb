```python
from sklearn import datasets # iris 데이터
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
```

```python
# 1. 데이터
iris = datasets.load_iris()
iris_X = iris.data
iris_y = iris.target
train_X, test_X,train_y,test_y = train_test_split(iris_X,iris_y,test_size=0.1, random_state=4,stratify=iris_y)

train_X.shape, test_X.shape,train_Y.shape,test_Y.shape
```

```python
# 사이킷런에서 제공하는 다층퍼셉트론 기반의 분류 알고리즘 api
mlp = MLPClassifier(hidden_layer_sizes=(60,50,30),
                    activation='relu',
                    solver='adam',# optimizer 설정값
                    max_iter=1000,
                    early_stopping=True,
                    n_iter_no_change=10, # 검증셋 손실함수가 개선되지 않으면 조기종료
                    validation_fraction=0.1)
                    
```

```python
mlp.fit(train_X,train_Y)
```

```python
mlp.score(test_X,test_Y)
```

```python
mlp.predict(test_X[0:1])
```

# 3. 클래스를 생성하여 DNN을 구현

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
```

```python
class DNNClassifier:
    @staticmethod
    def build(input_dim=4,activation='relu',output_dim=3):
#         model= Sequential()
#         model.add(Input(input_dim,))
#         model.add(Dense(50,activation=activation))
#         model.add(Dense(30,activation=activation))
#         model.add(Dense(output_dim,activation='softmax'))
        
        model = Sequential([
            Input(input_dim,),
            Dense(50,activation=activation),
            Dense(30,activation=activation),
            Dense(output_dim,activation='softmax'),
        ])
        # 학습과정 설정
        model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
        return model
        
```

```python
# 1. 데이터 생성 및 전처리
#iris_Y = pd.get_dummies(iris_y).to_numpy()
#train_X, test_X,train_Y,test_Y = train_test_split(iris_X,iris_Y,test_size=0.1, random_state=4,stratify=iris_y)
train_Y = pd.get_dummies(train_y).values
test_Y = pd.get_dummies(test_y).values
# 조기 종료 콜백
earlystopping = EarlyStopping(patience=150)
# 2. 모델
model = DNNClassifier.build(input_dim=4,activation='relu',output_dim=3)

# 3. 모델 학습
hist = model.fit(train_X,train_Y,
                 epochs= 1000,
                 validation_split=0.1,
                 callbacks=[earlystopping]
                )
```

```python
# 입력 데이터를 2차원으로 변환
input_data = test_X[0].reshape(1, -1)

# 예측 결과
prediction = model.predict(input_data)

# argmax는 예측 결과에서 가장 높은 확률을 가진 클래스 인덱스를 반환
predicted_class = prediction.argmax(axis=1)
print(predicted_class)
```

# 4. 함수 api 이용: 병렬처리

```python
from tensorflow.keras.layers import concatenate
from tensorflow.keras import Model
```

```python
train_X.shape, test_X.shape,train_Y.shape,test_Y.shape
```

```python
# 모델
input_ = Input(shape=(4,))
dense1 = Dense(50, activation='relu',)(input_)
dense2 = Dense(80, activation='relu',)(input_)
dense3 = Dense(30, activation='relu',)(input_)
x = concatenate([dense1,dense2,dense3])
x = Dense(units = 160, activation='relu')(input_)
#구현
output = Dense(units=3, activation='softmax')(x)
model = Model(inputs=input_, outputs=output)
model.summary()
```

```python
# 학습 정의(설정), 학습
model.compile(loss='categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
model.fit(train_X, train_Y,
         epochs=200,
         verbose=1)

```

```python

# 학습평가
loss, accuracy = model.evaluate(test_X, test_Y)
print(accuracy*100,'%')
```
